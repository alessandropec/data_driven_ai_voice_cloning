# Import path and parameters definition

```python
import torch
import torch.nn as nn

from hello.deepWorld.models.spresnet import SpResNet,resNet_prepro
from hello.deepWorld.models.sprecurrent import SpRecurrent,recurrent_to_spectr_prepro
import hello.deepWorld.trainingTools as trainingTools

from tqdm import tqdm

import hello.deepWorld.utils as deepUtils


import hello.dataWorld.dataset.pangramDataLoaders as panDL
from hello.dataWorld.audioTools import printAudio
from matplotlib.pyplot import imshow
from hello.dataWorld.dataset.pangramDataLoaders import pad_collate

import os
import json

```

# Train SpResNet on pangram dataset

```python
params=deepUtils.Params("./train_configs/pangram_sprecurrent_lstm.json")
print(json.load(open("./train_configs/pangram_sprecurrent_lstm.json")))

device = deepUtils.get_best_device()

model=SpRecurrent(
                    num_classes=params.num_classes,
                    input_size=params.input_size, 
                    num_layers=params.num_layers,
                    hidden_size=params.hidden_size,
                    embedding_size=params.embedding_size,
                    bidir=params.bidir,
                    dropout=params.dropout,
                    device=device,
                    model_type=params.model_type
                )

train_loader,_=panDL.get_SpeakerSignalDataLoader(files_dir=params.data_dir+"/train",signalType="wav",
                                    prepro_func=recurrent_to_spectr_prepro,batch_size=params.batch_size,num_workers=params.num_workers,collate_fn=pad_collate)
val_loader,_=panDL.get_SpeakerSignalDataLoader(files_dir=params.data_dir+"/val",signalType="wav",
                                    prepro_func=recurrent_to_spectr_prepro,batch_size=params.batch_size,num_workers=params.num_workers,collate_fn=pad_collate)

loss_fn = nn.CrossEntropyLoss()


optimizer = torch.optim.Adam(model.parameters(), lr=params.lr, weight_decay=params.weight_decay)

if params.scheduler:
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, 25, gamma=0.1)
else:
    scheduler = None



trainingTools.train_and_evaluate_accuracy(model,train_loader=train_loader,val_loader=val_loader,
                    optimizer=optimizer,device=device,loss_fn=loss_fn,scheduler=scheduler,params=params)
```

# Continue training

```python
params=deepUtils.Params("./train_configs/pangram_sprecurrent_gru.json")
paramsPrint=json.load(open("./train_configs/pangram_sprecurrent_gru.json"))

device = deepUtils.get_best_device()

model=SpRecurrent(
                    num_classes=params.num_classes,
                    input_size=params.input_size, 
                    num_layers=params.num_layers,
                    hidden_size=params.hidden_size,
                    embedding_size=params.embedding_size,
                    bidir=params.bidir,
                    dropout=params.dropout,
                    device=device,
                    model_type=params.model_type
                )
ckpt=torch.load(os.path.join(params.checkpoint_dir,"last.pth.tar"))
last_epoch=ckpt["epoch"]
model.load_state_dict(ckpt["model"])
model.to(device)


optimizer=torch.optim.Adam(model.parameters())
optimizer.load_state_dict(ckpt["optimizer"])


train_loader,_=panDL.get_SpeakerSignalDataLoader(files_dir=params.data_dir+"/train",signalType="wav",
                                    prepro_func=recurrent_to_spectr_prepro,batch_size=params.batch_size,
                                    num_workers=params.num_workers,collate_fn=pad_collate)
val_loader,_=panDL.get_SpeakerSignalDataLoader(files_dir=params.data_dir+"/val",signalType="wav",
                                    prepro_func=recurrent_to_spectr_prepro,batch_size=params.batch_size,
                                    num_workers=params.num_workers,collate_fn=pad_collate)

loss_fn = nn.CrossEntropyLoss()

if params.scheduler:
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, 25, gamma=0.1)
else:
    scheduler = None


print(f"Parameters used:\n {paramsPrint}")
print(f"Model:\n {model}")
print(f"\nLast epoch: {last_epoch}")
print(f"\nDevice: {device}")
print(f"\nOptimizer: {optimizer}")

print("Initial evaluation:\n")
deepUtils.evaluate(model=model,device=device,test_loader=val_loader,loss_fn=loss_fn)
```
