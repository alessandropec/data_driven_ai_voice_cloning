```python
from dataset import Dataset
from torch.utils.data import DataLoader
import yaml
from utils.model import get_model
from utils.tools import to_device

# Read Config
preprocess_config = yaml.load(
    open("config/LibriTTS/preprocess.yaml", "r"), Loader=yaml.FullLoader
)

train_config = yaml.load(open("config/LibriTTS/train.yaml", "r"), Loader=yaml.FullLoader)


model_config = yaml.load(open("config/LibriTTS/model.yaml", "r"), Loader=yaml.FullLoader)

configs = (preprocess_config, model_config, train_config)
```

```python
# Get dataset
dataset = Dataset(
    "train.txt", preprocess_config, train_config, sort=True, drop_last=True
)
batch_size = train_config["optimizer"]["batch_size"]
group_size = 1  # Set this larger than 1 to enable sorting in Dataset
assert batch_size * group_size < len(dataset)
loader = DataLoader(
    dataset,
    batch_size=batch_size * group_size,
    shuffle=True,
    collate_fn=dataset.collate_fn,
)
```

```python
import torch
class Struct:
    def __init__(self, **entries):
        self.__dict__.update(entries)

args=Struct(**{"restore_step":0})
model, optimizer = get_model(args, configs, torch.device("cuda"), train=True)
```

```python
for batchs in loader:
    for batch in batchs:
            batch=to_device(batch,"cuda",preprocess_config["speaker_emb"])
            output = model(*(batch[2:]))
            break
    break

batch
```
